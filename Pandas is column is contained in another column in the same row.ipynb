## Search if a column is part of another column on the same row

```python
movies['rat'] = movies.content_rating.astype(str)
```

```python
movies.head()
```

## Using lambda expression

* **Axis 0** iterate on all the ROWS in each COLUMN
* **Axis 1** iterate on all the COLUMNS in each ROW

```python
df_filter = movies[movies.apply(lambda row: row.rat in row.movie_title, axis=1)]
df_filter[['content_rating','movie_title']].head()
```

```python
df_filter = movies[movies.apply(lambda row: row.movie_title.find(row.rat) != -1, axis=1)]
df_filter[['content_rating','movie_title']].head()
```

## Using for loop and series to search

```python
rating_list = movies['content_rating']
title_list = movies['movie_title']
```

```python
is_part = []
for i, rate in enumerate(rating_list):
    if not pd.isna(title_list[i]) and (str(rate) in title_list[i]):
        is_part.append(True)
    else:
        is_part.append(False)
is_part[:5] 
```

```python
movies[is_part][['content_rating','movie_title']].head()
```

## search part of a column it it contained in another column?

```python
keywords_list = movies['plot_keywords']
title_list = movies['movie_title']
keys_list = movies['plot_keywords'].str.split('|').str.get(0)
keys_list[:5]
```

```python
title_list[:5]
```

```python
is_part = []
for i, key in enumerate(keys_list):
    if not pd.isna(title_list[i]) and (str(key) in title_list[i]):
        is_part.append(True)
    else:
        is_part.append(False)
is_part[:5]        
```

```python
movies[is_part][['plot_keywords','movie_title']]
```

```python
is_part = []
for i, key in enumerate(keys_list):
    if not pd.isna(title_list[i]) and (str(key) in title_list[i] or str(key).lower() == title_list[i].strip().lower()):
        is_part.append(True)
    else:
        is_part.append(False)
is_part
```

```python
movies[is_part][['plot_keywords','movie_title']]
```

## Performance tests

* for loop and comparison - 8.304 seconds
* lambda expresion - 25.893 seconds

```python
import cProfile
import pandas as pd
movies = pd.read_csv('./csv/movie_metadata.csv')
movies['rat'] = movies.content_rating.astype(str)


def before(movies):
    for i in range(1, 100):
        is_part = []
        title_list = movies['movie_title']
        genre_list = movies['content_rating']
        for i, genre in enumerate(genre_list):
            if str(genre) in title_list[i]:
                is_part.append(True)
            else:
                is_part.append(False)
        movies[is_part]


def after(movies):
    for i in range(1, 100):
        movies[movies.apply(lambda x: x.rat in x.movie_title, axis=1)]


cProfile.run("before(movies)")
cProfile.run("after(movies)")
```
